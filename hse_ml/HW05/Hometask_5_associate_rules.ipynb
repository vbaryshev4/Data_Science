# Домашнее задание 5. Поиск ассоциативных правил
**Задача**: Ваша исследовательская задача - понять, есть ли причинно-следственные связи между диагнозами сердечных заболеваний.


**Данные**: https://archive.ics.uci.edu/ml/datasets/spect+heart . Данные содержат 22 бинарных признака - 21 вспомогательный диагноз и 1 финальный для каждого пациента


**Дедлайн: 23:59 13.12.2018**

```python
import pandas as pd
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/spect/SPECT.train', header=None, 
                 names=['final'] + ['diag_{}'.format(i) for i in range(21)])
df.head()
```

```python
cols = list(df.columns)
```

```python
cols
```

```python
for i in cols:
    df[i] = df[i].replace(1, 'True_' + i)
    df[i] = df[i].replace(0, 'False_' + i)
```

```python
df.head()
```

```python
import pyfpgrowth
```

```python
patterns = pyfpgrowth.find_frequent_patterns(df.values, 50) 
rules = pyfpgrowth.generate_association_rules(patterns, 0.85)
```

```python
patterns
```

```python
rules
```

```python
for i in rules:
    if 'False_final' in i:
        print(i, rules[i], '\n')
```

```python
for i in rules:
    if 'True_final' in i or 'True_final' in rules[i]:
        print(i, rules[i], '\n')
```
