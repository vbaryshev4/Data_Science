```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats as st
```

```python
# TODO: Определить необходимое число наблюдений 
#       в эксперименте для получения значимого результата:
# Текущий уровень конверсии: 1%
# Минимальный уровень изменений: +0.1%
# Уровень значимости: 95%
# Требуемая мощность критерия: 90%
```

```python
n = np.linspace(1, 200001, 1000)
```

```python
p_0 = 0.01
p_1 = p_0 + 0.001
```

```python
SE = np.sqrt(p_1*(1-p_1))/n #формула стандартного отклонения для пропорций
```

```python
z = st.norm.ppf(0.95)
```
