```python
def convert_string_jsons(df, jsons_columns=['device', 'geoNetwork', 'totals', 'trafficSource']):
    for column in jsons_columns:
        df[column] = df[column].apply(lambda x: json.loads(x))

convert_string_jsons(train_df)
convert_string_jsons(test_df)
```

```python
def exctract(x, key):
    try:
        return x[key]
    except KeyError:
        return 'no_key'


def flatten(df):
    jsons_columns = [column for column in df.columns if type(df[column].values[0]) == dict]    
    
    if len(jsons_columns) > 0:
        for column in tqdm_notebook(jsons_columns):
            keys_set = set()
            for v in df[column].values:
                keys_set = keys_set.union(v.keys())
            for key in list(keys_set):
                print(key)
                df[key] = [exctract(x, key) for x in df[column]]
        df.drop(jsons_columns, axis=1, inplace=True)
        flatten(df)
        
    else:
        return None
```

```python
# Prints columns and df name if more than one type in column
# Returns columns name list with doubles in types 
def itter_types(df, dfname):

    def check_types(df, column):
        types = []
        for elem in df[column]:
            if type(elem) not in types:
                types.append(type(elem))
        return types

    result = []
    for column in df:
        r = check_types(df, column)
        if len(r) > 1:
            result.append(column)
            print('MORE THAN ONE TYPES IN {}'.format(dfname), column, r)
    return result

train_column_list_doubles = itter_types(train_df, 'train_df')
test_column_list_doubles = itter_types(test_df, 'test_df')
```

```python
def merge_types(df, column_name):
    values = []
    
    for elem in df[column_name]:
        if column_name == 'fullVisitorId':
            if isinstance(elem, int):
                values.append(str(elem))
            else:
                values.append(elem)
        if column_name == 'isTrueDirect':
            if isinstance(elem, bool):
                values.append(str(elem))
            else:
                values.append(elem)
        if column_name == 'isVideoAd':
            if isinstance(elem, bool):
                values.append(str(elem))
            else:
                values.append(elem)
        if column_name == 'targetingCriteria':
            if isinstance(elem, dict):
                values.append('empty_dict')
            else:
                values.append(elem)
            
    df[column_name] = pd.Series(values)

doubles = ['fullVisitorId', 'isTrueDirect', 'isVideoAd', 'targetingCriteria']

for double in doubles:
    merge_types(train_df, double)
    merge_types(test_df, double)
```

#### Get datas from pickles

```python
%%time
train_df = pd.read_pickle('train_df.pickle')
train_df['date'] = pd.to_datetime(train_df['date'], format='%Y%m%d')
train_df['weekday'] = train_df['date'].dt.weekday_name
train_df['day_off'] = (train_df['date'].dt.dayofweek > 4).astype(int)
```

```python
%%time
test_df = pd.read_pickle('test_df.pickle')
test_df['date'] = pd.to_datetime(test_df['date'], format='%Y%m%d')
test_df['weekday'] = test_df['date'].dt.weekday_name
test_df['day_off'] = (test_df['date'].dt.dayofweek > 4).astype(int)
```

```python
cat_features = ['channelGrouping','isMobile','browser','deviceCategory','operatingSystem',
               'country','region','subContinent','networkDomain','continent','newVisits',
               'medium', 'isTrueDirect', 'source', 'isVideoAd', 'slot','page', 'day_off', 'weekday']
num_features = ['visitNumber','pageviews','hits']
```

```python
#11) У фичей заменить редкие значения на others или -1
# Гипотеза прорвалилась, либо даст пр итонкой настрорйке лучший результат
shape = train_df.shape[0]

for column in tqdm_notebook(cat_features):
    train_df[column].value_counts()
    value_counts = train_df[column].value_counts() / shape
    anomalies = list(value_counts[value_counts < 0.01].keys())
    not_anomalies = list(value_counts[~(value_counts < 0.01)].keys())
    replace_dict = dict()

    for value in anomalies:
        if type(value) == str:
            replace_dict[value] = 'other'
        else:
            replace_dict[value] = -1
    for value in not_anomalies:
        replace_dict[value] = value
    train_df[column] = train_df[column].map(replace_dict)
    test_df[column] = test_df[column].map(replace_dict)
```

```python
shape = train_df.shape[0]
X_train, X_test = train_df.iloc[:int(shape*0.8),:], train_df.iloc[int(shape*0.8):,:]
y_train, y_test = X_train['transactionRevenue'], X_test['transactionRevenue']

X_train, X_test = X_train[cat_features+num_features], X_test[cat_features+num_features]

X_train[num_features] = X_train[num_features].replace('no_key',0)
X_test[num_features] = X_test[num_features].replace('no_key',0)
```

```python
categorical_features=[i for i, x in enumerate(X_train.columns) if x in cat_features]
```

```python
test_df = test_df[['fullVisitorId']+cat_features+num_features]
test_df[num_features] = test_df[num_features].replace('no_key',0).fillna(0)
test_df[cat_features] = test_df[cat_features].fillna('other')
```

# Start CatBoost

```python
from catboost import CatBoostRegressor
```

```python
model_Cat = CatBoostRegressor(iterations=100, thread_count=4, learning_rate=0.3,
                             depth=6, border_count=100, has_time=True,
                             counter_calc_method='SkipTest')
```

```python
model_Cat.fit(X_train, y_train, cat_features=categorical_features)
```

```python
for val, name in zip(model_Cat.feature_importances_, model_Cat.feature_names_):
    print(val, name)
```

```python
test_df['prediction'] = model_Cat.predict(test_df.iloc[:,1:])
test_df['prediction'] = test_df['prediction']*(test_df['prediction'] >= 0)
test_df['prediction'] = np.exp(test_df['prediction']) - 1
pred_dict = dict(test_df.groupby('fullVisitorId')['prediction'].sum())
```

```python
subm = pd.read_csv('sample_submission.csv')
subm['PredictedLogRevenue'] = subm['fullVisitorId'].map(pred_dict).fillna(0)
subm['PredictedLogRevenue'] = np.log(subm['PredictedLogRevenue']+1)
subm.to_csv('third_subm.csv', sep=',', index=None)
```
